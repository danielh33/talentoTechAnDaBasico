## 10. Sentencia HAVING

```python
"""Ejemplo: Análisis de goles por temporada en las ligas de España, Francia
e Inglaterra."""
query = """SELECT season,
            Country.name AS Country,
            League.name AS League,
            ROUND(AVG(Match.home_team_goal),2) AS avg_local_goals,
            ROUND(AVG(Match.away_team_goal),2) AS avg_visitor_goals,
            SUM(home_team_goal+away_team_goal) AS total_goals
        FROM Match
        JOIN Country ON Country.id = Match.country_id
        JOIN League ON League.id = Match.league_id
        WHERE Country.name IN ('Spain', 'France', 'England')
        GROUP BY season, League
        HAVING avg_local_goals BETWEEN 1.4 AND 1.6 AND avg_visitor_goals BETWEEN 1.2 AND 1.6
        ORDER BY season DESC;"""

goals_by_season = pd.read_sql(query, conn)
goals_by_season
```

```python
"""Práctica: Nos basaremos en ejemplo 'Lista detallada de juegos', pero
eliminaremos la información del partido y fechas, para comparar las ligas
por temporadas"""
query = """
SELECT C.NAME AS COUNTRY,
  L.NAME LEAGUE,
  COUNT(DISTINCT STAGE) STAGES,
  SEASON,
  COUNT(DISTINCT TH.team_long_name) +
    COUNT(DISTINCT TA.team_long_name) NUMBER_OF_TEAMS,
  ROUND(AVG(HOME_TEAM_GOAL), 2) avg_local_goals,
  ROUND(AVG(AWAY_TEAM_GOAL), 2) avg_visitor_goals,
  ROUND(AVG(HOME_TEAM_GOAL) + AVG(AWAY_TEAM_GOAL), 2) avg_goals,
  ROUND(AVG(HOME_TEAM_GOAL) - AVG(AWAY_TEAM_GOAL), 2) avg_goal_dif,
  SUM(HOME_TEAM_GOAL + AWAY_TEAM_GOAL) total_goals
FROM Country C
  JOIN Match M ON C.id = M.country_id
  JOIN LEAGUE L ON L.id = M.league_id
  JOIN Team TH ON TH.team_api_id = M.home_team_api_id
  JOIN Team TA ON TA.team_api_id = M.AWAY_team_api_id
WHERE C.NAME IN ('Spain', 'Germany', 'France', 'Italy', 'England')
GROUP BY 1,2,4
HAVING AVG_GOALS > 2.5
"""

#No modificar lo siguiente:
leages_by_season = pd.read_sql(query, conn)
leages_by_season
```

## 12. Sentencia CASE WHEN en SELECT

```python
"""Ejemplo: Queremos una tabla que me indique los partidos jugados por el FC
Barcelona, su rival, la fecha de juego y si perdió, ganó o empato."""
query= """
SELECT DATE(Match.date) as MatchDate,
  Home.team_long_name AS  HomeTeam,
  home_team_goal AS Home_goals,
  Visitor.team_long_name AS VisitorTeam,
  away_team_goal Visitor_goals,
  CASE
      WHEN Home.team_api_id = 8634 AND home_team_goal < away_team_goal
        THEN 'Barca lost :('
      WHEN Home.team_api_id = 8634 AND home_team_goal > away_team_goal
        THEN 'Barca won'
      WHEN Visitor.team_api_id = 8634 AND home_team_goal < away_team_goal
        THEN 'Barca won'
      WHEN Visitor.team_api_id = 8634 AND home_team_goal > away_team_goal
        THEN 'Barca lost :('
      ELSE 'Tie'
  END AS Outcome
FROM Match
LEFT JOIN Team AS Home on Home.team_api_id = Match.home_team_api_id
LEFT JOIN Team AS Visitor on Visitor.team_api_id = Match.away_team_api_id
WHERE (Visitor.team_api_id = 8634 OR Home.team_api_id = 8634)
  AND season = "2012/2013"
Order by MatchDate DESC
LIMIT 15;"""

Barca_results = pd.read_sql(query, conn)
Barca_results
```

##13. SUB QUERIES

```python
"""Ejemplo: Rating de Cristiano Ronaldo y Lionel Messi
versus los demás jugadores."""
query = """
SELECT CASE
    WHEN player_name = 'Cristiano Ronaldo' THEN 'Cristiano Ronaldo'
    WHEN player_name = 'Lionel Messi' THEN 'Lionel Messi'
    ELSE 'Others'
  END AS Player,
PA_Grouped.avg_overall_rating AS avg_overall_rating,
PA_Grouped.avg_potential AS avg_potential
FROM Player
  LEFT JOIN (
    SELECT Player_Attributes.player_api_id,
          ROUND(avg(Player_Attributes.overall_rating),2) AS avg_overall_rating,
          ROUND(avg(Player_Attributes.potential),2) AS avg_potential
      FROM Player_Attributes
      GROUP BY Player_Attributes.player_api_id
  ) AS PA_Grouped
  ON Player.player_api_id = PA_Grouped.player_api_id
GROUP BY Player
ORDER BY Player;"""

players_rating = pd.read_sql(query, conn)
players_rating
```
