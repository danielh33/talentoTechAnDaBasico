#**Taller SQL**
Vamos a analizar un cojunto de datos de las ligas europeas de futbol.

```python
# Import libraries
import pandas as pd
import sqlite3

from google.colab import drive # Traemos la clase drive, dentro del módulo colab, del paquete google
drive.mount('/content/drive') # Establecemos la conexión através del método mount

# Los datos de entrada están disponibles en el directorio "../input/".

path = "/content/drive/MyDrive/TALENTO_TECH_ANALISIS_DATOS_BASICO/practica3/"
database = path + 'database.sqlite'

# Creamos la conexión con la BD y listaremos las tablas con las que contamos

conn = sqlite3.connect(database)
print ("Conexión exitosa")

tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", conn)
tables
```

```python
query = '''
SELECT *
FROM Country;
'''
countries = pd.read_sql(query, conn)
countries
```

```python
# consulta todas las ligas del futbol europeo
query = '''
SELECT *
FROM League;
'''
leagues = pd.read_sql(query, conn)
leagues
```

```python
# Colsuta todos los jusgadores del futbol europeo
query = '''
SELECT *
FROM Player;
'''
players = pd.read_sql(query, conn)
players
```

```python
# Consultar los diferentes nombre de los jugadores del futbol europeo
query = '''
SELECT DISTINCT player_name
FROM Player;
'''
players_name = pd.read_sql(query, conn)
players_name
```

```python
# Consultar las diferentes temporadas del futbol europeo
query = '''SELECT DISTINCT season
FROM Match;
'''
seasons = pd.read_sql(query, conn)
seasons
```
